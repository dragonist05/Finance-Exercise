Some rolling stats calculations

```python
def rolling_stats_calculator(sym):

 stats = pd.DataFrame(dat[sym])
 window = 20
 stats["min"] = stats[sym].rolling(window=window).min()
 stats["max"] = stats[sym].rolling(window=window).max()
 stats["std"] = stats[sym].rolling(window=window).std()
 stats["mean"] = stats[sym].rolling(window=window).mean()
 stats["median"] = stats[sym].rolling(window=window).median()
 return stats
```

```python
bitcoin_stats = rolling_stats_calculator("Bitcoin")
ethereum_stats = rolling_stats_calculator("Ethereum")
xrp_stats = rolling_stats_calculator("XRP")
cardano_stats = rolling_stats_calculator("Cardano")
```

Plotting our rolling stats

```python
fig, axes = plt.subplots(2, 2, figsize=(20,15))
axes = axes.flatten()

crypto_data = [
    (bitcoin_stats, "Bitcoin"),
    (ethereum_stats, "Ethereum"),
    (xrp_stats, "XRP"),
    (cardano_stats, "Cardano"),
]

for i, (data, sym) in enumerate(crypto_data):
    ax = axes[i]

    data[['min', 'mean', 'max']].iloc[-200:].plot(
        ax=ax,
        style=['g--', 'r--', 'g--'], 
        lw=0.8,
        title = f"{sym} stats"
    )
    
    data[sym].iloc[-200:].plot(ax=ax, lw=2.0, color='b')
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', labelrotation=40)



plt.tight_layout()
plt.show()
```

Calculating log returns and plotting them on a scatter matrix

```python
rets = np.log(dat / dat.shift(1))
rets.head()
```

```python
pd.plotting.scatter_matrix(rets, 
 alpha=0.5, 
 diagonal='hist', 
 hist_kwds={'bins': 35}, 
 figsize=(15, 10))
```

Let's implement OLS regression and make a scatter plot 

```python
sym_list = [ "Ethereum", "XRP", "Cardano" ]
fig, axes = plt.subplots(1,3)
rets = rets.dropna()

for i,sym in enumerate(sym_list):
    ax = axes[i]
    reg = np.polyfit(rets[sym], rets["Bitcoin"], deg=1)

    rets.plot(
        ax = ax,
        kind="scatter",
        x = sym,
        y = "Bitcoin",
        color = "b",
        figsize=(15,10)
    )
    ax.plot(rets[sym], np.polyval(reg, rets[sym]), c = "r", lw=2)



    
plt.tight_layout()
plt.show()
```

Lastly let's do some correlation using log returns and plot them

```python
rets.corr()
```

Plotting rolling and static correlations

```python
fig, axes = plt.subplots(3,1)
axes.flatten()

#calculating their correlation to Bitcoin
for i, sym in enumerate(sym_list):
    ax = axes[i]
    rets["Bitcoin"].rolling(window=200).corr(rets[sym]).plot(
        ax = ax,
        color = "r",
        lw = 2.0,
        title = f"Bitcoin and {sym} correlation",
        figsize=(15,10)
    )
    ax.axhline(rets.corr().iloc[0,(i+1)], c="y")
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis="x", labelrotation = 40)

plt.tight_layout()
plt.show()
```
